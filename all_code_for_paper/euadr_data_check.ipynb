```python
def convert(text):
    mapping = {
        "PA": "positive",
        "NA": "negative",
        "SA": "speculative",
        "FA": "false"
    }
    
    assert text in mapping, "{0}".format(text)
    return mapping[text]
```

Read all of the files and check that the number of relations matches what was published

```python
def get_toby_id(subset, concept_a, concept_b, vals):
    """
    For one raw annotated relationship, determine which of the
    published relationships it matches.
    
    Ignores all errors at the moment...
    """
    if concept_a[3] == "Chemicals & Drugs": # disease first
        concept_a, concept_b = concept_b, concept_a
        
#     print "inside get toby id"
        
#     print subset
#     print
#     print concept_a
#     print concept_b
        
    ans = set()
        
    def match(entity, concept, row):
        name = row["{0}_TEXT".format(entity)]
        
        if name == concept[0]: # exact name match:
            return True
        
        if len(concept[0]) < len(name) and concept[0] in name:
            return True
        
        return False
    
    assert vals[1] in ["True", "False"]
    
    for index, row in subset.iterrows():
        if match("ENTITY1", concept_a, row) and match("ENTITY2", concept_b, row):
            if vals[1] == "True":
                reltype = convert(vals[10])
            else:
                reltype = "false"
                
            if reltype == row["ASSOCIATION_TYPE"]:
                ans.add(row["toby_id"])

                
                
    if len(ans) == 1:
        return list(ans)[0]
    
    # deal with the errors here later
    return ""
```

```python
loc = "euadr_corpus_complete"
bad_papers = []
total = 0

num_experts_agree = dict()

for pmid, subset in pub_euadr.groupby("PMID"):
    fname = "{0}.txt".format(pmid)

    num_relations = len(subset)

    print "File:", fname
    
    print "Has {0} published relations".format(num_relations)
    
#     print
#     print subset
#     print
    
    relations = 0
    concepts = dict()
    
    for line in read_file(fname, loc):
        vals = line.split("\t")
        if vals[0] != "Drug-Disorder":
            break
            
        assert len(vals) in [10, 11]        
        
        if len(vals) == 10: # concept
            assert vals[9] in ["Diseases & Disorders", "Chemicals & Drugs"]
            concepts[vals[8]] = (vals[3], vals[4], vals[5], vals[9]) # name, start, stop, type
        else: # drug disease relation
            annotators = vals[9].split(',')
            
            humans = set(annotators) - set(["Computer"])
            
            if len(humans) >= 2:
                # create this relation
                relation_id = get_toby_id(subset, concepts[vals[3]], concepts[vals[4]], vals)
                
                if relation_id:
                    num_experts_agree[relation_id] = len(humans)
                    relations += 1
                    
    print "found {0}".format(relations)
                
    if relations != num_relations:
        bad_papers.append(pmid)
        
    total += relations
        
    print
        
        
print "total", total
        
        
        
print "num papers", num_papers
print len(bad_papers)
print bad_papers
            
            
        
    
print "done"
```

```python
with open("num_experts_agree.txt", "w") as out:
    for toby_id, num in num_experts_agree.items():
        out.write("{0}|{1}\n".format(toby_id, num))
```
