```python
def aggregate_votes(data):
    """
    Given all of the human responses for one work unit,
    aggregates the results for the broad semantic choice.

    For each choice (positive, speculative, negative,
    false), calculates: total number of votes, confidence
    score, and the percentage agreement. Sorts results
    in descending order of confidence score.

    Confidence score is the sum of the accuracy score
    of the individual participants.
    """
    temp = []
    for correlation_type, group in data.groupby("sentence_claim"):
        conf_score = sum(group["_trust"])
        num_votes = len(group)
        temp.append([correlation_type, conf_score, num_votes])

    return pd.DataFrame(temp, columns = ["sentence_claim", "conf_score", "num_votes"])
```
