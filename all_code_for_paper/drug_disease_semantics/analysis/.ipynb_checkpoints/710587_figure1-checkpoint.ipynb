```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%pylab inline
```

```python
import sys
sys.path.append("/home/toby/crowdsourcing/drug_disease_semantics/src/")

from filter_data import filter_data
from true_relation_type import get_euadr_relation_type
from broad_reltype_performance import determine_broad_reltype_results

from broad_reltype_summary import broad_reltype_summary
from get_orig_problem import get_orig_problem

from collections import defaultdict
```

```python
def num_correct(settings):
    raw_data = filter_data(settings)
    true_relation_type = get_euadr_relation_type(settings)
    results = determine_broad_reltype_results(true_relation_type, raw_data)
    
    num_right = 0
    for unit_id, group in results.groupby("unit_id"):
        num_right += group.iloc[0]["match"]
        
    return num_right
```

```python
categories = [
    [["positive"], ["false"], ["negative"], ["speculative"]],
    [["positive", "speculative"], ["false"], ["negative"]],
    [["positive"], ["speculative"], ["false", "negative"]],
    [["positive", "speculative"], ["false", "negative"]]
]
```

```python
ans = []
for combine in categories:
    settings = {
        "loc": "/home/toby/crowdsourcing/drug_disease_semantics/data",
        "fname": "job_710587_full_with_untrusted.tsv",
        "data_subset": "normal",
        "min_accuracy": 0.7,
        "max_accuracy": 1.0,
        "sentence_type": "all",
        "categories": combine
    }
    
    ans.append(num_correct(settings))
```

```python
ans
```

```python
labels = [
    "Strict match,\n no choices combined",
    "Positive and\n speculative combined",
    "False and \n negative combined",
    "Positive and \nspeculative combined,\n false and negative combined"
]

percent_match = pd.Series(ans, index = labels) / 60 * 100
print percent_match

graph = percent_match.plot(kind = "bar",
                           figsize = (10, 7),
                           title = "Percentage match with EU-ADR for different conditions",
                           rot = 1)
graph.set_xlabel("Different choices combined")
graph.set_ylabel("Percentage of sentences where\n the top crowd answer matches that of the EU-ADR")
```
