```python
import pandas as pd
import os
import sys
```

```python
loc = "drug_disease_semantics/data"
fname = "EUADR_drug_disease_with_toby_identifiers.tsv"

orig_data = pd.read_csv(os.path.join(loc, fname), sep = "\t")
```

```python
sys.path.append("/home/toby/crowdsourcing/drug_disease_semantics/src/")
from filter_data import filter_data
```

```python
settings = {
    "loc": "/home/toby/crowdsourcing/drug_disease_semantics/data",
    "fname": "job_710587_full_with_untrusted.tsv",
    "data_subset": "normal",
    "min_accuracy": 0.7,
    "max_accuracy": 1.0,
    "sentence_type": "all"
}

raw_data = filter_data(settings)
```

```python
mapping = dict()
for unit_id, group in raw_data.groupby("_unit_id"):
    assert len(group["pubmed_id"].unique()) == 1
    pmid = group.iloc[0]["pubmed_id"]
    subset = orig_data.query("PMID == {0}".format(pmid))
    
    assert len(group["subject_text"].unique()) == 1
    assert len(group["object_text"].unique()) == 1
    
    sub_text = group.iloc[0]["subject_text"]
    obj_text = group.iloc[0]["object_text"]
        
    num_sent = group.iloc[0]["num_sentence"]
    
    subset = subset.query("ENTITY1_TEXT == '{0}'".format(sub_text))
    subset = subset.query("ENTITY2_TEXT == '{0}'".format(obj_text))
    subset = subset.query("NUM_SENTENCE == {0}".format(num_sent))
    
    assert group.iloc[0]["subject_start"] == subset.iloc[0]["ENTITY1_INI"]
    assert group.iloc[0]["object_start"] == subset.iloc[0]["ENTITY2_INI"]
    
    assert len(subset) == 1
    
    toby_id = subset.iloc[0]["toby_id"]
    
    mapping[unit_id] = toby_id
```

```python
with open("unit_id_mapping.txt", "w") as out:
    for unit_id, toby_id in mapping.items():
        out.write("{0}|{1}\n".format(unit_id, toby_id))
```
