Function for getting the raw relation list from the original annotator file

```python
def convert(text):
    mapping = {
        "PA": "positive",
        "SA": "speculative",
        "NA": "negative",
        "FA": "false"
    }
    assert text in mapping
    return mapping[text]

def determine_reltype(final, computer):
    assert final in ["True", "False"]
    if final == "True":
        return convert(computer)
    
    return "false"

def get_raw_relation_list(pmid):
    fname = "id_{0}.txt".format(pmid)
    loc = "id_raw_euadr_corpus"
    
    concepts = dict()
    relation = dict()
    
    for line in read_file(fname, loc):
        vals = line.split('\t')
        assert len(vals) in [10, 12]
        if len(vals) == 10:
            if vals[8] not in concepts:
                concepts[vals[8]] = (vals[3], vals[4], vals[5], vals[9]) # name, start, stop, type
            else:
                # some concepts are annotated twice
                # it seems like the computer gets it wrong, and then the humans had to correct it
                
                # need to replace?
                if vals[1] == "True":
                    concepts[vals[8]] = (vals[3], vals[4], vals[5], vals[9]) # name, start, stop, type
        else:
            rel_id = vals[11]
            
            sub = vals[3]
            obj = vals[4]
            
            if sub not in concepts or obj not in concepts:
                # some of the raw annotations refer to concepts which do not exist
                # example: pmid 17634547
                # rel_5 is between concepts 16 and 14, but concept 14 doesn't exist...
                continue
            
            
            
            # keep it in disease, drug order
            if concepts[sub][3] == "Chemicals & Drugs":
                sub, obj = obj, sub
                
            reltype = determine_reltype(vals[1], vals[10])
            
            num_agree = len(set(vals[9].split(",")) - set(["Computer"]))
            final = vals[1]
            
            relation[rel_id] = (reltype, num_agree, concepts[sub], concepts[obj], final)
            
    return relation
```

```python
def concept_match(sent_idx, concept_start, concept_stop, sent_start, sent_stop):
    """
    Determines if a concept is at the right spot in this sentence/abstract.
    
    concept: index of concept with respect to abstract
    sent: index of concept with respect to sentence
    """
    sent_idx = int(sent_idx)
    concept_start = int(concept_start)
    concept_stop = int(concept_stop)
    sent_start = int(sent_start)
    sent_stop = int(sent_stop)
    
    right_start = concept_start == (sent_idx + sent_start)
    right_stop = concept_stop == (sent_idx + sent_stop)
    
    return right_start and right_stop
```

```python
def find_match(row, raw_relations, abstract):
    """
    Find a match between the toby_id of the published euadr relations
    and the raw relation id, using the indicies of the pubmed paper.
    """
    sentence = row["SENTENCE"]
    assert sentence in abstract
    
    
    sent_idx = abstract.index(sentence)
    
    # look for row through all raw_relations
    matches = set()
    for rel_id, info in raw_relations.items():
        sub = info[2]
        obj = info[3]
        
        if (concept_match(sent_idx, sub[1], sub[2], row["ENTITY1_INI"], row["ENTITY1_END"])
            and concept_match(sent_idx, obj[1], obj[2], row["ENTITY2_INI"], row["ENTITY2_END"])):
            matches.add(rel_id)
            
    if len(matches) == 1:
        return list(matches)[0]
    
    if len(matches) > 1:
        ans = set()
        # look through and find the "True" one
        for rel_id in matches:
            if raw_relations[rel_id][4] == "True":
                ans.add(rel_id)
                
        if len(ans) == 1:
            return list(ans)[0]
        
        # couldn't find exactly one "true" answer (eg they're all false??)
        
        assert len(ans) == 0
        
        # either too many or none
        # look for the one with most agreement
        most_agree = -1
        for rel_id in matches:
            most_agree = max(most_agree, raw_relations[rel_id][1])

        temp = set()
        for rel_id in matches:
            if int(raw_relations[rel_id][1]) == most_agree:
                temp.add(rel_id)

                
        # 17894421, rel_5 and 11, two diff people both agreed with diff answers for the real ans
        # speculative and negative
        # since num agree is same, we can just randomly pick one... (figure this out later)
        
        return list(temp)[0]
    
#         assert len(temp) == 1, temp
#         return list(temp)[0]
        
    
    # try an exact name match... (desperate measures)
    matches = set()
    for rel_id, info in raw_relations.items():
        sub = info[2]
        obj = info[3]
        
        if sub[0] == row["ENTITY1_TEXT"] and obj[0] == row["ENTITY2_TEXT"]:
            matches.add(rel_id)
            
    assert len(matches) == 1, matches
    return list(matches)[0]
    
    raise Exception("no match!")
```

```python
# all sentences can be found directly..
mapping = dict()
for pmid, group in pub_euadr.groupby("PMID"):
    print "Pubmed id:", pmid
    
    title, chunks = get_abstract_information(str(pmid))
    
    abstract = title + " " + " ".join(chunks)
    
    raw_relations = get_raw_relation_list(pmid)
    
    for index, row in group.iterrows():
        raw_id = find_match(row, raw_relations, abstract)
        
        mapping[row["toby_id"]] = (raw_id, raw_relations[raw_id][1])
        
print "done"

header = ["toby_id", "raw_relation_id", "num_expert_agree"]
with open("raw_to_toby_mapping.txt", "w") as out:
    out.write("{0}\n".format("|".join(header)))
    for toby_id, info in mapping.items():
        out.write("{0}|{1}|{2}\n".format(toby_id, info[0], info[1]))
    
```
