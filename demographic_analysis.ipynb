This notebook will look at the demographic data collected about the participants of CrowdFlower job #710587. Since we are unable to determine the participants' age, sex, or education status, we will only look at what countries the participants live in.

```python
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pycountry

%matplotlib inline
%pylab inline
```

```python
from src.filter_data import filter_data
```

### Here is the function for generating our plot:

```python
def plot_countries_of_origin(settings):
    raw_data = filter_data(settings)
    
    users = raw_data[["_worker_id", "_country"]].drop_duplicates()
    print "Number of participants: {0}".format(len(users))
    
    # group participants together by country
    names = []
    nums = []
    for code, num in Counter(users["_country"]).most_common():
        country = pycountry.countries.get(alpha3 = code)

        names.append(country.name.split(',')[0]) # remove things like ', republic of'
        nums.append(num)
    
    # create the plot
    res = pd.Series(nums, index = names)
    graph = res.plot(kind = 'barh', figsize = settings["figsize"],
                     title = 'Country of origin of participants',
                     fontsize = '12')

    graph.set_xlabel("Number of participants")
```

### Let's first look at all the participants, regardless of whether they worked on the actual task:

```python
settings = {
    "loc": "data",
    "fname": "job_710587_full_with_untrusted_id.tsv",
    "data_subset": "all",
    "min_accuracy": 0.0,
    "max_accuracy": 1.0,
    "categories": [["positive"], ["false"], ["negative"], ["speculative"]],
    "figsize": (5, 15)
}

plot_countries_of_origin(settings)
```

### Now let's look at the participants who worked on the actual task:

```python
settings = {
    "loc": "data",
    "fname": "job_710587_full_with_untrusted_id.tsv",
    "data_subset": "normal",
    "min_accuracy": 0.7,
    "max_accuracy": 1.0,
    "categories": [["positive"], ["false"], ["negative"], ["speculative"]],
    "figsize": (6, 6)
}

plot_countries_of_origin(settings)
```
